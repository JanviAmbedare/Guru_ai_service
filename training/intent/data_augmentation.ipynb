```python
import random
import pandas as pd

# ======================================================
# ADVANCED AUGMENTATION RULES
# ======================================================

PREFIXES = [
    "",
    "please",
    "guru",
    "hey guru",
    "can you",
    "could you",
    "would you",
    "guru please",
    "hey",
    "listen guru"
]

SUFFIXES = [
    "",
    "please",
    "right now",
    "immediately",
    "for me",
    "quickly"
]

HINGLISH_VARIATIONS = {
    "move forward": [
        "aage chalo",
        "aage jao"
    ],

    "stop": [
        "ruk jao",
        "bas karo"
    ],

    "turn left": [
        "left lo",
        "baaye jao"
    ],

    "turn right": [
        "right lo",
        "daaye jao"
    ]
}

TYPO_VARIATIONS = {
    "forward": [
        "forword",
        "foward"
    ],

    "battery": [
        "battry",
        "battary"
    ],

    "reminder": [
        "remnder",
        "remainder"
    ]
}

ELDERLY_PATTERNS = [
    "beta",
    "please help",
    "can you help me",
    "i need assistance",
    "please do this"
]
```

```python

def clean_text(text):

    return (
        text.lower()
        .strip()
    )


# TYPO GENERATOR

def add_typo_variation(text):

    words = text.split()

    updated = []

    for word in words:

        if (
            word in TYPO_VARIATIONS and
            random.random() < 0.3
        ):

            updated.append(
                random.choice(
                    TYPO_VARIATIONS[word]
                )
            )

        else:

            updated.append(word)

    return " ".join(updated)


# HINGLISH AUGMENTATION
def add_hinglish(text):

    if text in HINGLISH_VARIATIONS:

        return random.choice(
            HINGLISH_VARIATIONS[text]
        )

    return text


# ELDERLY SPEECH
def add_elderly_style(text):

    prefix = random.choice(
        ELDERLY_PATTERNS
    )

    return f"{prefix} {text}"


# MAIN AUGMENTATION
def augment_text(text):

    augmented = set()

    augmented.add(text)

    for prefix in PREFIXES:

        for suffix in SUFFIXES:

            variant = (
                f"{prefix} {text} {suffix}"
            ).strip()

            augmented.add(
                clean_text(variant)
            )

    augmented.add(
        add_typo_variation(text)
    )

    augmented.add(
        add_hinglish(text)
    )


    augmented.add(
        add_elderly_style(text)
    )

    return list(augmented)
```

```python

# DATASET GENERATION

def generate_dataset(base_data):

    rows = []

    for intent, phrases in base_data.items():

        for phrase in phrases:

            augmented_phrases = (
                augment_text(phrase)
            )

            for augmented in augmented_phrases:

                rows.append({

                    "text": clean_text(
                        augmented
                    ),

                    "intent": intent
                })

    return pd.DataFrame(rows)
```

```python
# SAVE DATASET


df = generate_dataset(base_data)

df = df.drop_duplicates()

df.to_csv(

    "../../training/intent/augmented_dataset.csv",

    index=False
)

print(df.head())

print(
    f"\nTotal samples: {len(df)}"
)
```
